```python
import os
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset

class LGGDataset(Dataset):
    """
    Custom Dataset class for LGG Brain Tumor Segmentation.
    This class:
    - Loads MRI images
    - Loads corresponding tumor masks
    - Resizes images
    - Normalizes pixel values
    - Converts everything into PyTorch tensors
    """

    def __init__(self, root_dir):
        """
        root_dir: Path to kaggle_3m folder
        Example: "lgg_data/kaggle_3m"
        """
        self.images = []
        self.masks = []

        # Walk through all patient folders
        for root, dirs, files in os.walk(root_dir):
            for file in files:

                # Select only image files (not mask files)
                if file.endswith(".tif") and "_mask" not in file:

                    image_path = os.path.join(root, file)
                    mask_path = os.path.join(root, file.replace(".tif", "_mask.tif"))

                    # Store image and mask paths
                    self.images.append(image_path)
                    self.masks.append(mask_path)

    def __len__(self):
        """
        Returns total number of image-mask pairs
        """
        return len(self.images)

    def __getitem__(self, idx):
        """
        Loads one image-mask pair
        idx: index of the sample
        """

        # Load MRI image in grayscale
        image = cv2.imread(self.images[idx], cv2.IMREAD_GRAYSCALE)

        # Load corresponding tumor mask
        mask = cv2.imread(self.masks[idx], cv2.IMREAD_GRAYSCALE)

        # Resize both image and mask to fixed size (256x256)
        # Neural networks require consistent dimensions
        image = cv2.resize(image, (256, 256))
        mask = cv2.resize(mask, (256, 256))

        # Normalize image pixel values (0–255 → 0–1)
        image = image / 255.0

        # Convert mask to binary (0 or 1)
        mask = mask / 255.0
        mask = (mask > 0.5).astype(np.float32)

        # Add channel dimension (1, 256, 256)
        # Required because CNN expects (C, H, W)
        image = np.expand_dims(image, axis=0)
        mask = np.expand_dims(mask, axis=0)

        # Convert numpy arrays to PyTorch tensors
        image = torch.tensor(image, dtype=torch.float32)
        mask = torch.tensor(mask, dtype=torch.float32)

        return image, mask
```

```python
# Create dataset object
dataset = LGGDataset("lgg_data/kaggle_3m")

# Print total samples
print("Total samples:", len(dataset))
```

```python
from torch.utils.data import DataLoader, random_split

# Define split ratio
train_ratio = 0.8

# Calculate sizes
train_size = int(train_ratio * len(dataset))
val_size = len(dataset) - train_size

# Split dataset randomly
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# Create DataLoaders
# DataLoader handles batching and shuffling

train_loader = DataLoader(
    train_dataset,
    batch_size=8,       # Number of images per batch
    shuffle=True        # Shuffle training data for better learning
)

val_loader = DataLoader(
    val_dataset,
    batch_size=8,
    shuffle=False       # No need to shuffle validation data
)

# Print sizes
print("Training samples:", len(train_dataset))
print("Validation samples:", len(val_dataset))
```

# **Building U-Net Model**

```python
import torch
import segmentation_models_pytorch as smp

# Check if GPU is available (important for faster training)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


# -------------------------------------------------------
# Create U-Net model
# -------------------------------------------------------
# encoder_name="resnet34"
#   → Pretrained feature extractor (transfer learning)
#
# encoder_weights="imagenet"
#   → Uses weights trained on ImageNet dataset
#
# in_channels=1
#   → MRI images are grayscale (1 channel)
#
# classes=1
#   → Binary segmentation (tumor vs background)
# -------------------------------------------------------

model = smp.Unet(
    encoder_name="resnet34",
    encoder_weights="imagenet",
    in_channels=1,
    classes=1,
).to(device)


# -------------------------------------------------------
# Define Loss Function
# -------------------------------------------------------
# Dice Loss is best for medical segmentation because:
# - Handles class imbalance
# - Focuses on overlap between prediction and ground truth

loss_fn = smp.losses.DiceLoss(mode="binary")


# -------------------------------------------------------
# Define Optimizer
# -------------------------------------------------------
# Adam optimizer adjusts learning rate automatically

optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)


print("Model created successfully!")
```

# **Defining** Dice Metric (Evaluation Metric)

```python
# -------------------------------------------------------
# Dice Score Function
# Measures overlap between prediction and ground truth
# Higher is better (range: 0 to 1)
# -------------------------------------------------------

def dice_score(pred, target, smooth=1e-5):

    # Convert logits → probabilities
    pred = torch.sigmoid(pred)

    # Convert probabilities → binary mask
    pred = (pred > 0.5).float()

    # Compute intersection
    intersection = (pred * target).sum()

    # Compute Dice coefficient
    dice = (2 * intersection + smooth) / \
           (pred.sum() + target.sum() + smooth)

    return dice
```

```python
# -------------------------------------------------------
# Training Loop
# -------------------------------------------------------

epochs = 5   # You can increase later (10–20 for better accuracy)

for epoch in range(epochs):

    # ---------------------
    # TRAINING PHASE
    # ---------------------
    model.train()
    train_loss = 0

    for images, masks in train_loader:

        # Move data to GPU
        images = images.to(device)
        masks = masks.to(device)

        # Forward pass
        outputs = model(images)

        # Compute loss
        loss = loss_fn(outputs, masks)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss += loss.item()


    # ---------------------
    # VALIDATION PHASE
    # ---------------------
    model.eval()
    val_dice = 0

    with torch.no_grad():   # No gradient calculation during validation

        for images, masks in val_loader:

            images = images.to(device)
            masks = masks.to(device)

            outputs = model(images)

            val_dice += dice_score(outputs, masks).item()


    # ---------------------
    # Print Results
    # ---------------------
    print(f"Epoch [{epoch+1}/{epochs}]")
    print(f"Train Loss: {train_loss/len(train_loader):.4f}")
    print(f"Validation Dice: {val_dice/len(val_loader):.4f}")
    print("-"*40)
```

```python
model.eval()

images, masks = next(iter(val_loader))
images = images.to(device)
masks = masks.to(device)

with torch.no_grad():
    outputs = model(images)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("MRI")
plt.imshow(images[0][0].cpu(), cmap="gray")

plt.subplot(1,3,2)
plt.title("Ground Truth")
plt.imshow(masks[0][0].cpu(), cmap="gray")

plt.subplot(1,3,3)
plt.title("Prediction")
plt.imshow(torch.sigmoid(outputs[0][0]).cpu(), cmap="gray")

plt.show()
```

```python
# -------------------------------------------------------
# Calculate Tumor Volume for one validation image
# -------------------------------------------------------

model.eval()

# Get one batch from validation set
images, masks = next(iter(val_loader))

images = images.to(device)

with torch.no_grad():
    outputs = model(images)

# Convert prediction to binary mask
pred_mask = torch.sigmoid(outputs[0])
pred_mask = (pred_mask > 0.5).float()

# Count tumor pixels
tumor_pixels = pred_mask.sum().item()

# Assume each pixel = 1 mm^3 (for demo)
voxel_volume = 1  # mm³

tumor_volume = tumor_pixels * voxel_volume

print("Tumor Pixel Count:", tumor_pixels)
print("Estimated Tumor Volume (mm³):", tumor_volume)
print("Estimated Tumor Volume (cm³):", tumor_volume / 1000)
```
